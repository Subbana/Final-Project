```python
import pandas as pd
import requests
import json
```

```python
df = pd.read_csv("data.csv")
df.head()
```

```python
api_key = "3f5d994574cfd43b79141f91f3d85b32"
base = "http://api.walkscore.com/score?format=json"
```

```python
def get_score(lat, lon):
    url = f"{base}&lat={lat}&lon={lon}&wsapikey={api_key}"
    score = requests.get(url).json()["walkscore"]
    return score
```

```python
scores = []
for _ in range(df.shape[0]):
    lat = df.latitude[_]
    lon = df.longitude[_]
    try:
        scores.append(get_score(lat, lon))
    except:
        scores.append("N/A")
```

```python
len(scores)
```

```python
df["walk_score"] = pd.Series(scores).values
```
