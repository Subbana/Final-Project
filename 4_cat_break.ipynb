```python
import pandas as pd
```

```python
df = pd.read_csv("final_data.csv")
df.head()
```

```python
df1 = df.copy()
def cat1(categories):
    cat1 = categories.split("'")[1]
    if cat1 != "foodtrucks":
        return cat1
df1["Category"] = df1["Categories"].map(cat1)
```

```python
df2 = df.copy()
def cat2(categories):
    try:
        cat2 = categories.split("'")[3]
    except:
        cat2 = None
    if cat2 != "foodtrucks":
        return cat2
df2["Category"] = df2["Categories"].map(cat2)
```

```python
df3 = df.copy()
def cat3(categories):
    try:
        cat3 = categories.split("'")[5]
    except:
        cat3 = None
    if cat3 != "foodtrucks":
        return cat3
df3["Category"] = df3["Categories"].map(cat3)
```

```python
df_cleaned = pd.concat([df1, df2, df3])
df_cleaned.head()
```

```python
df_cleaned.info()
```

```python
df_cleaned.to_csv("cat_data.csv", index=False)
```
